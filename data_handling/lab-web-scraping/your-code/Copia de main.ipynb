####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python

eq = requests.get('https://www.emsc-csem.org/Earthquake/')
eq
````

````python
eq_cont=eq.content
soup_7=bs(eq_cont, 'html.parser')
soup_7
````

````python

table_eq = soup_7.findAll('tr',{'class':'normal'},limit=20)

date = []
time=[]
lat=[]
long=[]
reg=[]
for i in table_eq:
    date.append(re.findall(r'\d+\-\d+\-\d+',i.text)[0])
for i in table_eq:
    time.append(re.findall(r'\d+\:\d+\:\d+',i.text)[0])
for i in table_eq:
    lat.append(re.findall(r'\d+\.\d+\s\w',i.text)[0])
for i in table_eq:
    long.append(re.findall(r'\d+\.\d+\s\w',i.text)[0])
    
table_reg = soup_7.findAll('td',{'class':'tb_region'},limit=20)
for i in table_reg:
    reg.append(re.findall(r'[A-Z].+',i.text)[0])



    
print(date)
print(time)
print(lat)
print(long)
print(reg)

clean_lat=[i.replace('\xa0',' ') for i in lat]
clean_long=[i.replace('\xa0',' ') for i in lat]
print(clean_lat)
print(clean_long)

quakes={'DATE':date,'TIME':time,'LATITUDE':lat,'LONGITUDE':long,'REGION NAME':reg}
pd.DataFrame(quakes)

                 
                                                                     
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
wiki_lang = requests.get('https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers')
wiki_lang
````

````python
wiki_lang_cont=wiki_lang.content
soup_10=bs(wiki_lang_cont, 'html.parser')
soup_10
````

````python
table_wiki_lang = soup_10.find('table',{'class':'wikitable sortable'})
headers=table_wiki_lang.findAll('th')
headers
head = [i.text.replace('\n','') for i in headers]
head

````

````python
rows = table_wiki_lang.find_all('tr')
rows = [i.text.strip().split('\n') for i in rows]
rows.pop(0)
rows2 = [[e for e in i if e != '']for i in rows]
rows2
##we want the top 10 so:
del rows2[10:]
rows2

````

````python
langs_table = pd.DataFrame(rows2, columns = head)
langs_table
clean = lambda x: re.sub(r'\[\d+\]','',x)
langs_table['Language']=langs_table['Language'].apply(clean)
langs_table
````
