```python
##checando las columnas que tienen valores vacios

null_cols = users_posts.isnull().sum()
null_cols[null_cols > 0]
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before passing to the next step

```python
##tomamos un parametro con null=0, para conocer el numero de datos 
full_col=users_posts['userId'].notnull().sum()
##con este valor se puede tener un mejor panorama, y se toma la desicion de conservar aquellas columnas que tengan un registro
##de al menos la mitad menos 10mil de nuestro parametro que cuenta con todos los registros

drop_cols = list(null_cols[null_cols > (full_col/2)+10000].index)
clean_users_posts = users_posts.drop(drop_cols, axis=1)
clean_users_posts.info()
```

```python
import numpy as np

##baja varianza
low_variance = []

for columna in clean_users_posts._get_numeric_data():
    minimum = min(clean_users_posts[columna])
    ninety_perc = np.percentile(clean_users_posts[columna], 90)
    if ninety_perc == minimum:
        low_variance.append(columna)

print(low_variance)

##no hay columnas que cumplan la condicion de la baja varianza
```

#### Bonus: Identify extreme values in your merged dataframe as you have learned in class, create a dataframe called outliers with the same columns as our data set and calculate the bounds. The values of the outliers dataframe will be the values of the merged_df that fall outside that bounds. You will need to save your outliers dataframe to a csv file on your-code folder.

```python
import numpy as np


out = clean_users_posts.describe().transpose()
out['IQR'] = out['75%']-out['25%']
out
```

```python
outliers = pd.DataFrame(columns=clean_users_posts.columns)

for col in out.index:
    iqr = out.at[col,'IQR']
    cutoff = iqr * 15
    lower = out.at[col,'25%'] - cutoff
    upper = out.at[col,'75%'] + cutoff
    results = clean_users_posts[(clean_users_posts[col] < lower) | 
                   (clean_users_posts[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)

outliers.head()
```

```python

outliers.to_csv('outliers.csv')
```
