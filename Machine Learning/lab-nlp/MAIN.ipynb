```python
txt="@Ironhack's-#Q website 776-is http://ironhack.com [(2018)])"
```

```python
import re
def clean_up(s):
    c=re.sub('[^a-zA-Z]|http\S+', ' ', s.lower())
    return c.strip()

```

```python
clean=clean_up(txt)
clean
```

```python
def tokenize(t):
    return word_tokenize(t)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
def stem_and_lemmatize(l):
    lst = []
    lemmatizer = WordNetLemmatizer()
    porter = PorterStemmer()
    for p in l:
        lst.append(lemmatizer.lemmatize(porter.stem(p)))
    return lst
```

```python
verbs_lst=['fucking','eaten','runs','said','killed']
stem_and_lemmatize(verbs_lst)
```
